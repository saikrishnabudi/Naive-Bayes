```python
# Accuray Score

classifier_mb=MB()
classifier_mb.fit(x,y)
test_pred_m=classifier_mb.predict(x)
accuracy_test_m=np.mean(test_pred_m==y)

print("Testing Accuracy is:",accuracy_test_m*100)
```
