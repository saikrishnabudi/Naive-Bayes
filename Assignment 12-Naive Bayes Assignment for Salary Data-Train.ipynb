```python
salary1=salary_train.iloc[:,0:13]
salary1=pd.get_dummies(salary1)
salary1
```

```python
final_salary=pd.concat([salary1,salary_train["Salary"]],axis=1)
final_salary
```

```python
x=final_salary.values[:,0:102]
y=final_salary.values[:,102]
```

```python
x
```

```python
# Accuracy Score

classifier_mb=MB()
classifier_mb.fit(x,y)
train_pred_m=classifier_mb.predict(x)
accuracy_train_m=np.mean(train_pred_m==y)

print("The Training Accuracy is:",accuracy_train_m*100)
```
